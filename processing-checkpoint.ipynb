### Import data

```python
# Find files with the .pickle extension in a file.
def find_pickle_files(data_folder):
  pickle_files = []
  for root, dirs, files in os.walk(data_folder):
    for file in files:
      if file.endswith(".pickle"):
        pickle_files.append(os.path.join(root, file))
  return pickle_files
```

```python
# Read the data in the found .pickle file.
def get_pickle_data(data_links):
    pickle_data = {}  
    for file in data_links:
        try:
            with open(file, "rb") as f:
                data = pickle.load(f)
            pickle_data[file] = data
        except (FileNotFoundError, PermissionError) as e:
            print(f"ERROR {file}: {e}")
    return pickle_data
```

```python
historical_folder = "data/historical_data/"
historical_data_links = find_pickle_files(historical_folder)
historical_dict = get_pickle_data(historical_data_links)
```

# Data Preparation

### Analyze race detail table (RA)

```python
race_detail = historical_dict["data/historical_data/RA.pickle\\RA.pickle"]
race_detail.head(3)
```

### Select the columns of data that can be used (RA)

```python
race_detail=race_detail[["id$Year","id$MonthDay","id$JyoCD","id$Kaiji","id$Nichiji","id$RaceNum","Kyori","TrackCD","TenkoBaba$DirtBabaCD","TenkoBaba$SibaBabaCD","TenkoBaba$TenkoCD"]]
race_detail.head(3)
```

```python
# Clear empty data rows
race_detail=race_detail.dropna(axis=0, how='any', subset=None, inplace=False)
```

 ### Make column: race_id (RA)

```python
# combine "id$Year","id$MonthDay", "id$JyoCD", "id$Kaiji", "id$Nichiji", "id$RaceNum" cols and add "id" column to the dataset
race_id_cols = ["id$Year", "id$MonthDay", "id$JyoCD", "id$Kaiji", "id$Nichiji", "id$RaceNum"]
race_id_df = race_detail[race_id_cols]

race_id_df = race_id_df.astype("str")

race_id_df["id$MonthDay"] = race_id_df["id$MonthDay"].apply(lambda x: "0"+x if len(x) == 3 else x)
race_id_df["id$Nichiji"] = race_id_df["id$Nichiji"].apply(lambda x: "0"+x if len(x) == 1 else x)
race_id_df["id$RaceNum"] = race_id_df["id$RaceNum"].apply(lambda x: "0"+x if len(x) == 1 else x)

# add race_id to race_detail dataset
race_detail["race_id"] = race_id_df["id$Year"] + race_id_df["id$MonthDay"] + race_id_df["id$JyoCD"] \
                    + race_id_df["id$Kaiji"] + race_id_df["id$Nichiji"] + race_id_df["id$RaceNum"]
```

### Race detail table (RA)

```python
# id: race id
# Kyori: total distance (meter)
# TrackCD: track code, type of racetrack
# TenkoBaba$SibaBabaCD: turf condition. 0: unknown, 1: firm, 2: good, 3: yielding, 4: soft
# TenkoBaba$DirtBabaCD: dirt condition. 0: unknown, 1: standard, 2: good, 3: muddy, 4: sloppy
# TenkoBaba$TenkoCD: Weather code. 0: unknown, 1: Fine, 2: Cloudy,3: Rainy, 4: Drizzle, 5: Snow, 6: Light Snow

race_info_cols = ["race_id", "Kyori", "TrackCD","TenkoBaba$TenkoCD", "TenkoBaba$SibaBabaCD", "TenkoBaba$DirtBabaCD"]

race_df = race_detail[race_info_cols]

race_df.rename(columns={"Kyori": "distance(m)", "TrackCD": "racetrack_type","TenkoBaba$TenkoCD":"weather_type",
                        "TenkoBaba$SibaBabaCD": "turf_condition", "TenkoBaba$DirtBabaCD": "dirt_condition"}, inplace=True)
```

### Analyze Horse Information (SE)

```python
info_detail = historical_dict["data/historical_data/SE.pickle\\SE.pickle"]
info_detail.head(5)
```

### Select the columns of data that can be used (SE)

```python
info_detail=info_detail[["id$Year", "id$MonthDay", "id$JyoCD", "id$Kaiji", "id$Nichiji", "id$RaceNum","Umaban", "KettoNum", "KisyuCode", "BanusiCode", "ChokyosiCode", 
                   "BaTaijyu","ZogenFugo","ZogenSa", "Futan", "SexCD", "Barei", "HinsyuCD", "KakuteiJyuni","Time"]]
```

```python
info_detail.head(5)
```

### Calculate the weight of a horse

Columns: BaTaijyu, ZogenFugo, ZogenSa

```python
info_detail["ZogenSa"] = info_detail["ZogenSa"].apply(lambda x: "000" if x=="   " else x)
info_detail[["ZogenSa", "BaTaijyu"]] = info_detail[["ZogenSa", "BaTaijyu"]].astype("int32")
info_detail["BaTaijyu"] = info_detail["BaTaijyu"].apply(lambda x: info_detail["BaTaijyu"].median() if (x == 999 or x == 0) else x)
```

```python
def weight_calc(original_weight, sign, weight_change):
    if sign == "+":
        weight = original_weight + weight_change
    elif sign == "-":
        weight = original_weight - weight_change
    else:
        weight = original_weight
    return weight
info_detail["horseweight"] = 0
info_detail["horseweight"] = info_detail.apply(
    lambda x: weight_calc(x["BaTaijyu"], x["ZogenFugo"], x["ZogenSa"]), axis=1)
```

### Make column: race_id (SE)

```python
# combine "id$Year","id$MonthDay", "id$JyoCD", "id$Kaiji", "id$Nichiji", "id$RaceNum" cols and add "id" column to the dataset
horse_id_cols = ["id$Year", "id$MonthDay", "id$JyoCD", "id$Kaiji", "id$Nichiji", "id$RaceNum"]
horse_id_df = info_detail[horse_id_cols]

# cast raceetrack_num from str to int32
horse_id_df["id$JyoCD"] = horse_id_df["id$JyoCD"].astype("int32")

horse_id_df = horse_id_df.astype("str")

horse_id_df["id$MonthDay"] = horse_id_df["id$MonthDay"].apply(lambda x: "0"+x if len(x) == 3 else x)
horse_id_df["id$JyoCD"] = horse_id_df["id$JyoCD"].apply(lambda x: "0"+x if len(x) == 1 else x)
horse_id_df["id$Nichiji"] = horse_id_df["id$Nichiji"].apply(lambda x: "0"+x if len(x) == 1 else x)
horse_id_df["id$RaceNum"] = horse_id_df["id$RaceNum"].apply(lambda x: "0"+x if len(x) == 1 else x)

info_detail["race_id"] = horse_id_df["id$Year"] + horse_id_df["id$MonthDay"] + horse_id_df["id$JyoCD"] \
                    + horse_id_df["id$Kaiji"] + horse_id_df["id$Nichiji"] + horse_id_df["id$RaceNum"]
```

### Horse Infomation Table

```python
# Umaban: horse number in a race
# id$JyoCD: racetrack code
# Futan: carry weight
# SexCD: horse gender. 1: male, 2: female, 3: castrated
# Barei: horse age
# HinsyuCD: product code, types of horses
# KisyuCode: jockeys' code
# BanusiCode: owners' code
# ChokyosiCode: trainers' code
# KakuteiJyuni: result

# Get data that is suitable for prediction
info_cols = ["Umaban", "race_id", "KettoNum", "KisyuCode", "BanusiCode", "ChokyosiCode", 
                   "horseweight", "Futan", "SexCD", "Barei", "HinsyuCD", "KakuteiJyuni", "Time"]

info_df = info_detail[info_cols]

info_df.rename(columns={"Umaban": "horse_num", "KettoNum": "horse_id",
                         "KisyuCode": "jockey_code", "BanusiCode": "owner_code", "ChokyosiCode": "trainer_code", 
                        "Futan": "carry_weight", "Barei": "horse_age",
                        "HinsyuCD": "horse_type", "KakuteiJyuni": "result"}, inplace=True)
print(info_df.shape)
info_df.head()
```

## Calculate win rate top3 of jockeys, owners, trainers

Calculate winrate top3 of jockeys

```python
# 1. calculate the top 3 finish rate of each jockey
result_ratio = info_df[info_df['result'] <= 3].groupby('jockey_code').size() / info_df.groupby('jockey_code').size()
result_ratio = result_ratio.reset_index(name='jokey_rate_top3')
result_ratio['jokey_rate_top3'].fillna(0, inplace=True)

# 2. calculate the average result for each jockey
result_mean = info_df.groupby('jockey_code')['result'].mean().reset_index(name='jokey_avg_rank')

# 3. Add two new columns to the original DataFrame
info_df = pd.merge(info_df, result_ratio, on='jockey_code', how='left')
info_df = pd.merge(info_df, result_mean, on='jockey_code', how='left')
```

Calculate winrate top3 of owners

```python
# 1. calculate the top 3 finish rate of each owner
result_ratio = info_df[info_df['result'] <= 3].groupby('owner_code').size() / info_df.groupby('owner_code').size()
result_ratio = result_ratio.reset_index(name='owner_rate_top3')
result_ratio['owner_rate_top3'].fillna(0, inplace=True)

# 2. calculate the average result for each owner
result_mean = info_df.groupby('owner_code')['result'].mean().reset_index(name='owner_avg_rank')

# 3. Add two new columns to the original DataFrame
info_df = pd.merge(info_df, result_ratio, on='owner_code', how='left')
info_df = pd.merge(info_df, result_mean, on='owner_code', how='left')
```

Calculate winrate top3 of trainers

```python
# 1. calculate the top 3 finish rate of each trainer
result_ratio = info_df[info_df['result'] <= 3].groupby('trainer_code').size() / info_df.groupby('trainer_code').size()
result_ratio = result_ratio.reset_index(name='trainer_rate_top3')
result_ratio['trainer_rate_top3'].fillna(0, inplace=True)

# 2. calculate the average result for each trainer
result_mean = info_df.groupby('trainer_code')['result'].mean().reset_index(name='trainer_avg_rank')

# 3. Add two new columns to the original DataFrame
info_df = pd.merge(info_df, result_ratio, on='trainer_code', how='left')
info_df = pd.merge(info_df, result_mean, on='trainer_code', how='left')
```
